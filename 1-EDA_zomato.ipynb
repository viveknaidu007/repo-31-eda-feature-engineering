```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_val = final_df.Country.value_counts().values
```

```python
#now pie chart
#i want to see which coun try has maximu transactions

plt.pie(country_val,labels=country_names)
```

```python
#see above see bad names r jumbled
#now
#here i just taking top 3 country only

plt.pie(country_val[:3],labels=country_names[:3])
```

```python
#now we can see percentage using in the parameer autopct
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```
